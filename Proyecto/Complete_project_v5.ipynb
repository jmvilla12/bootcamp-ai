#### Eliminación de duplicados

```python
df_lluvia = df_lluvia.drop_duplicates(inplace=False)
df_rsolar = df_rsolar.drop_duplicates(inplace=False)
df_tseca = df_tseca.drop_duplicates(inplace=False)
```

Hay datos de la temperatura seca del aire cada 3 horas desde 1983-02-01 hasta 2020-01-31. Se tomó este dato ya que haciendo la respectiva investigación, ya que descubrimos que este es el dato generalmente usado para los pronosticos del clima en televisión y que muestran las aplicaciones

Todos los datos son tomados de la misma estación meteorológica, el aeropuerto internacional matecaña. Por lo tanto podemos ignorar muchos parámetros y enfocarnos en las variables como los campos ('VALOR') junto a sus respectivas fecha y hora para irlos clasificando

#### Se extraen las columnas de fecha y valor de cada dataframe que son las que nos interesan

```python
campos = ['Fecha', 'Valor']
diaria_lluvia = df_lluvia[campos].copy()
diaria_lluvia['Fecha'] = pd.to_datetime(diaria_lluvia['Fecha'])
diaria_lluvia.rename(columns={'Valor': 'Precipitacion'}, inplace=True)
diaria_lluvia['Fecha'] = pd.to_datetime(diaria_lluvia['Fecha']) #Convierto la fecha en tipo datetime
```

```python
new_rsolar = df_rsolar[campos].copy()
new_rsolar['Fecha'] = pd.to_datetime(new_rsolar['Fecha'])
new_rsolar
```

```python
# Como los datos de radiación solar están cada hora están por horas debemos sacar el promedio de radiación solar por dia
new_rsolar = new_rsolar[new_rsolar['Valor'] >= 5].copy() #Solo me interesan las horas reales de sol durante el dia, de noche se toman como cero
new_rsolar.loc[:, 'Dia'] = new_rsolar['Fecha'].dt.date #Extraigo la fecha sin hora en una nueva columna

diaria_rsolar = new_rsolar.groupby('Dia').agg(
    Radiacion_Promedio=('Valor', 'mean')  # Calcula el promedio diario
).reset_index()

diaria_rsolar.rename(columns={'Dia': 'Fecha'}, inplace=True)
diaria_rsolar['Fecha'] = pd.to_datetime(diaria_rsolar['Fecha']) #Convierto la fecha en tipo datetime
# print(diaria_rsolar)
```

```python
new_tseca = df_tseca[campos].copy()
new_tseca['Fecha'] = pd.to_datetime(new_tseca['Fecha'])
new_tseca
```

```python
new_tseca.loc[:, 'Dia'] = new_tseca['Fecha'].dt.date #Extraigo la fecha sin hora en una nueva columna
new_tseca

diaria_tseca = new_tseca.groupby('Dia').agg(
    Temperatura_Promedio=('Valor', 'mean')  # Calcula el promedio diario
).reset_index()

diaria_tseca.rename(columns={'Dia': 'Fecha'}, inplace=True)
diaria_tseca['Fecha'] = pd.to_datetime(diaria_tseca['Fecha']) #Convierto la fecha en tipo datetime
print(diaria_tseca)
```

```python
def crear_rango_fechas(df, columna_fecha, frecuencia='D'):
    '''
    Crea un rango de fechas diario a partir del dataframe

    Args:
      df: DataFrame con valores de fechas.
      columna_fecha: Nombre de la columna con las fechas.
      frecuencia: Frecuencia del rango de fechas (ej: 'D' diario, 'W' semanal)

    Returns:
      Un objeto pandas.DatetimeIndex con el rango de fechas.
    '''
    f_inicio = min(df[columna_fecha])
    f_fin = max(df[columna_fecha])
    rango_fechas = pd.date_range(start = f_inicio, end = f_fin, freq = frecuencia)
    return rango_fechas

rango_fechas_lluvia = crear_rango_fechas(diaria_lluvia, 'Fecha')
rango_fechas_rsolar = crear_rango_fechas(diaria_rsolar, 'Fecha')
rango_fechas_tseca = crear_rango_fechas(diaria_tseca, 'Fecha')

# Reconstruir el dataframe con todas las fechas
lluvia_completo = diaria_lluvia.set_index('Fecha').reindex(rango_fechas_lluvia).reset_index()
lluvia_completo.rename(columns={'index': 'Fecha'}, inplace=True)

rsolar_completo = diaria_rsolar.set_index('Fecha').reindex(rango_fechas_rsolar).reset_index()
rsolar_completo.rename(columns={'index': 'Fecha'}, inplace=True)

tseca_completo = diaria_tseca.set_index('Fecha').reindex(rango_fechas_tseca).reset_index()
tseca_completo.rename(columns={'index': 'Fecha'}, inplace=True)

# print(lluvia_completo)
# print(rsolar_completo)
# print(tseca_completo)

# print(lluvia_completo.isnull().sum())
# print(rsolar_completo.isnull().sum())
# print(tseca_completo.isnull().sum())
```

```python
fecha_inicio = '2014-12-08'
fecha_fin = '2022-09-11'
fecha_inicio = pd.to_datetime(fecha_inicio)
fecha_fin = pd.to_datetime(fecha_fin)


# Sólo gráficaremos después de fecha_inicio
lluvia_filtered = lluvia_completo[(lluvia_completo['Fecha'] >= fecha_inicio) & (lluvia_completo['Fecha'] <= fecha_fin)].copy().reset_index(drop=True)
rsolar_filtered = rsolar_completo[(rsolar_completo['Fecha'] >= fecha_inicio) & (rsolar_completo['Fecha'] <= fecha_fin)].copy().reset_index(drop=True)
tseca_filtered = tseca_completo[(tseca_completo['Fecha'] >= fecha_inicio) & (tseca_completo['Fecha'] <= fecha_fin)].copy().reset_index(drop=True)

# print(lluvia_filtered)
# print(rsolar_filtered)
# print(tseca_filtered)

tseca_filtered.info()
```

```python
# Combinar los tres DataFrames por la columna Fecha
df_filtered = pd.merge(lluvia_filtered, rsolar_filtered, on='Fecha', how='outer')
df_filtered = pd.merge(df_filtered, tseca_filtered, on='Fecha', how='outer')

# Identificar valores nulos
print(df_filtered.isnull().sum())
print(df_filtered)
```

```python
df_train = df_filtered.dropna()
print(df_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
# Entrenamos para predecir la precipitación
X_train = df_train[['Radiacion_Promedio', 'Temperatura_Promedio']]  # Variables independientes
y_train = df_train['Precipitacion']  # Variable dependiente

# Modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Revisar la importancia de las variables
print(f"Coeficientes del modelo: {modelo.coef_}")
```

```python
# Filtrar datos con precipitación nula pero radiación y temperatura completas
df_pred = df_filtered[df_filtered['Precipitacion'].isnull() & 
                      df_filtered['Radiacion_Promedio'].notnull() & 
                      df_filtered['Temperatura_Promedio'].notnull()]

# Variables para predicción
X_pred = df_pred[['Radiacion_Promedio', 'Temperatura_Promedio']]

# Predecir precipitación
df_filtered.loc[df_pred.index, 'Precipitacion'] = modelo.predict(X_pred)
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler
# Normalización
scaler = MinMaxScaler()
# scaler = RobustScaler()


# Normaliza cada dataset por separado
df_filtered.loc[:,'Precipitacion_Normalizada'] = scaler.fit_transform(df_filtered[['Precipitacion']])
df_filtered.loc[:,'Radiacion_Normalizada'] = scaler.fit_transform(df_filtered[['Radiacion_Promedio']])
df_filtered.loc[:,'Temperatura_Normalizada'] = scaler.fit_transform(df_filtered[['Temperatura_Promedio']])

plt.figure(figsize=(14, 6))

plt.plot(df_filtered['Fecha'], df_filtered['Precipitacion_Normalizada'], label='Precipitación (Normalizada)', color='blue', alpha=0.7)
plt.plot(df_filtered['Fecha'], df_filtered['Radiacion_Normalizada'], label='Radiación Solar (Normalizada)', color='orange', alpha=0.7)
plt.plot(df_filtered['Fecha'], df_filtered['Temperatura_Normalizada'], label='Temperatura (Normalizada)', color='green', alpha=0.7)

plt.title(f'Comparación Normalizada de Precipitación, Radiación Solar y Temperatura después ({fecha_inicio.date()} en adelante)', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Valor Normalizado', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
# Asegurarnos de que 'Fecha' esté en formato datetime en el dataset consolidado
df_filtered['Fecha'] = pd.to_datetime(df_filtered['Fecha'])

# Crear columna de semana basada en la primera fecha común
fecha_inicio = df_filtered['Fecha'].min()
df_filtered.loc[:, 'Semana'] = ((df_filtered['Fecha'] - pd.Timestamp(fecha_inicio)).dt.days // 7)

# Agrupar por semana y calcular los promedios para cada variable
semanal = df_filtered.groupby('Semana').agg({
    'Precipitacion_Normalizada': 'mean',
    'Radiacion_Normalizada': 'mean',
    'Temperatura_Normalizada': 'mean'
}).reset_index()

# Crear la columna de fecha inicial de cada semana
semanal['Fecha_Inicial'] = pd.to_datetime(fecha_inicio) + pd.to_timedelta(semanal['Semana'] * 7, unit='d')

# Unificar los datos en formato "largo" para graficar
semanal_largo = semanal.melt(
    id_vars=['Semana', 'Fecha_Inicial'],
    value_vars=['Precipitacion_Normalizada', 'Radiacion_Normalizada', 'Temperatura_Normalizada'],
    var_name='Variable',
    value_name='Promedio_Semanal'
)

# Opcional: Renombrar las variables para hacerlas más legibles en los gráficos
semanal_largo['Variable'] = semanal_largo['Variable'].replace({
    'Precipitacion_Normalizada': 'Precipitación',
    'Radiacion_Normalizada': 'Radiación Solar',
    'Temperatura_Normalizada': 'Temperatura'
})

semanal_largo
```

```python
# Graficar datos normalizados por semana
plt.figure(figsize=(14, 6))
for variable in semanal_largo['Variable'].unique():
    data = semanal_largo[semanal_largo['Variable'] == variable]
    plt.plot(data['Fecha_Inicial'], data['Promedio_Semanal'], label=variable, alpha=0.7)

plt.title('Comparación Normalizada Semanal de Variables', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Promedio Semanal Normalizado', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
