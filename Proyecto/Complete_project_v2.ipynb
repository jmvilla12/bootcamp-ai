## Fundamentación teórica
### 1. Fases del ciclo de vida de un proyecto de machine learning
El ciclo de vida de una aplicación de aprendizaje de máquina comienza con la identificación del problema, definiendo claramente el objetivo o problema y cómo el aprendizaje automático puede contribuir a resolverlo parcial o totalmente. Luego, se pasa a la recolección de datos, asegurando que sean relevantes, precisos, suficientes y alineados con el problema objetivo.

#### Definición del objetivo y analisis de viabilidad
La identificación del problema en Machine Learning es el primer paso. Implica definir claramente el desafío que se requiere solventar, entendiendo de primera mano su importancia y cómo el aprendizaje automático puede ser útil.

- Especificar el desafío de forma medible, es decir que se pretende predecir.
- Identificar y reconocer el valor y el impacto que la solución proporcionará en su aplicación.
- Asegurarse de que el problema es aplicable para técnicas de Machine Learning, es decir, si hay patrones en los datos que se pueden aprender para automatizar la toma de decisiones.
- Determinar si se trata de un problema de clasificación, regresión, clustering, etc.

#### Stakeholders

Todas aquellas personas, comunidades u organizaciones que se ven involucradas o afectadas en el propósito del proyecto de forma directa o indirecta.
identificar este tipo de tipología y gestionarlo eficazmente es crucial para asegurar que sus necesidades sean atendidas, no haya sesgos y que el proyecto sea exitoso en la medida de lo posible.

#### Acondicionamiento de datos

En la etapa de preparación de datos, se analizan, filtran, limpian, normalizan, completan y escalan los datos, además de manejar valores faltantes, indeterminados o atípicos, segmentarlos, transformarlos y dividirlos en conjuntos de entrenamiento y prueba. Posteriormente, en la ingeniería de modelos, se selecciona el modelo competente con la dificultad, el propósito y la capacidad de computo, se realizan ajustes de hiperparámetros y se entrena el modelo según criterios de desempeño que exige el objetivo principal.

#### Evaluación, métricas e indices de desempeño

Finalmente, en la evaluación del modelo, se miden métricas como precisión, recall y F1-score para determinar si el modelo es lo preciso, apto y acorde a lo esperado para ser desplegado.

#### Se extraen las columnas de fecha y valor de cada dataframe que son las que nos interesan

```python
campos = ['Fecha', 'Valor']
diaria_lluvia = df_lluvia[campos].copy()
diaria_lluvia['Fecha'] = pd.to_datetime(diaria_lluvia['Fecha'])
diaria_lluvia.rename(columns={'Valor': 'Precipitacion'}, inplace=True)
diaria_lluvia['Fecha'] = pd.to_datetime(diaria_lluvia['Fecha']) #Convierto la fecha en tipo datetime
```

```python
new_rsolar = df_rsolar[campos].copy()
new_rsolar['Fecha'] = pd.to_datetime(new_rsolar['Fecha'])
new_rsolar
```

```python
# Como los datos de radiación solar están cada hora están por horas debemos sacar el promedio de radiación solar por dia
new_rsolar = new_rsolar[new_rsolar['Valor'] >= 5].copy() #Solo me interesan las horas reales de sol durante el dia, de noche se toman como cero
new_rsolar.loc[:, 'Dia'] = new_rsolar['Fecha'].dt.date #Extraigo la fecha sin hora en una nueva columna

diaria_rsolar = new_rsolar.groupby('Dia').agg(
    Radiacion_Promedio=('Valor', 'mean')  # Calcula el promedio diario
).reset_index()

diaria_rsolar.rename(columns={'Dia': 'Fecha'}, inplace=True)
diaria_rsolar['Fecha'] = pd.to_datetime(diaria_rsolar['Fecha']) #Convierto la fecha en tipo datetime
# print(diaria_rsolar)
```

```python
new_tseca = df_tseca[campos].copy()
new_tseca['Fecha'] = pd.to_datetime(new_tseca['Fecha'])
new_tseca
```

```python
new_tseca.loc[:, 'Dia'] = new_tseca['Fecha'].dt.date #Extraigo la fecha sin hora en una nueva columna
new_tseca

diaria_tseca = new_tseca.groupby('Dia').agg(
    Temperatura_Promedio=('Valor', 'mean')  # Calcula el promedio diario
).reset_index()

diaria_tseca.rename(columns={'Dia': 'Fecha'}, inplace=True)
diaria_tseca['Fecha'] = pd.to_datetime(diaria_tseca['Fecha']) #Convierto la fecha en tipo datetime
print(diaria_tseca)
```

```python
def crear_rango_fechas(df, columna_fecha, frecuencia='D'):
    '''
    Crea un rango de fechas diario a partir del dataframe

    Args:
      df: DataFrame con valores de fechas.
      columna_fecha: Nombre de la columna con las fechas.
      frecuencia: Frecuencia del rango de fechas (ej: 'D' diario, 'W' semanal)

    Returns:
      Un objeto pandas.DatetimeIndex con el rango de fechas.
    '''
    f_inicio = min(df[columna_fecha])
    f_fin = max(df[columna_fecha])
    rango_fechas = pd.date_range(start = f_inicio, end = f_fin, freq = frecuencia)
    return rango_fechas

rango_fechas_lluvia = crear_rango_fechas(lluvia_filtered, 'Fecha')
rango_fechas_rsolar = crear_rango_fechas(rsolar_filtered, 'Fecha')
rango_fechas_tseca = crear_rango_fechas(tseca_filtered, 'Fecha')

# Reconstruir el dataframe con todas las fechas
lluvia_completo = lluvia_filtered.set_index('Fecha').reindex(rango_fechas_lluvia).reset_index()
lluvia_completo.rename(columns={'index': 'Fecha'}, inplace=True)

rsolar_completo = rsolar_filtered.set_index('Fecha').reindex(rango_fechas_rsolar).reset_index()
rsolar_completo.rename(columns={'index': 'Fecha'}, inplace=True)

tseca_completo = tseca_filtered.set_index('Fecha').reindex(rango_fechas_tseca).reset_index()
tseca_completo.rename(columns={'index': 'Fecha'}, inplace=True)

# print(lluvia_completo.isnull().sum())
# print(rsolar_completo.isnull().sum())
# print(tseca_completo.isnull().sum())
```

### Tratamiento de datos ausentes y anomalias
El manejo de datos faltantes es importante para que los valores nulos o vacíos no afecten la correlación de las métricas estadísticas, la interpretación, el análisis descriptivo y el desempeño del modelo de machine learning.

- Z-Score  
    - z= abs((x-media)/std)
- Media
- Mediana

```python
# Manejo de datos faltantes mediante la media de cada mes en función del mismo periodo de tiempo de años diferentes
rsolar_completo['Radiacion_Promedio'] = rsolar_completo['Radiacion_Promedio'].fillna(
    rsolar_completo.groupby(rsolar_completo['Fecha'].dt.month)['Radiacion_Promedio'].transform('mean')
)
# rsolar_completo['Radiacion_Promedio'] = rsolar_completo['Radiacion_Promedio'].interpolate(method='linear')

tseca_completo['Temperatura_Promedio'] = tseca_completo['Temperatura_Promedio'].fillna(
    tseca_completo.groupby(tseca_completo['Fecha'].dt.month)['Temperatura_Promedio'].transform('mean')
)

lluvia_completo['Precipitacion'] = lluvia_completo['Precipitacion'].interpolate(method='linear')
```

### 1. Validación de Distribución Normal
Este paso se usa para corroborar si los datos de una variable siguen una distribución normal (gaussiana).

#### Histograma
Leyendo los histogramas nos damos cuenta de que mientras la radiación solar y la temperatura tienen una distribución normal, la pluviosidad tiene una distribución sesgada positivamente.

```python
import seaborn as sns
sns.histplot(semanal_lluvia['Promedio_Semanal'],kde=True)
sns.histplot(semanal_rsolar['Promedio_Semanal'],kde=True)
sns.histplot(semanal_tseca['Promedio_Semanal'],kde=True)
```

#### Dispersión de los datos

```python
fig = plt.figure(figsize = (15, 6))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)
ax1.set_title('Precipitación')
ax1.plot(semanal_lluvia['Promedio_Semanal'], linewidth = 0, marker = 'o', color = 'blue', markersize = 6)
ax1.set_ylim(0,1)
ax2.set_title('Radiación solar')
ax2.plot(semanal_rsolar['Promedio_Semanal'], linewidth = 0, marker = 'o', color = 'orange', markersize = 6)
ax2.set_ylim(0,1)
ax3.set_title('Temperatura')
ax3.plot(semanal_tseca['Promedio_Semanal'], linewidth = 0, marker = 'o', color = 'green', markersize = 6)
ax3.set_ylim(0,1)
plt.show()
```

#### Gráfico Q-Q (Quantile-Quantile)
Compara los cuantiles de los datos en una distribución normal teórica. Los puntos deben alinearse a lo largo de una línea diagonal.

```python
import scipy.stats as stats
# Q-Q plot para las tres variable
fig, axs = plt.subplots(1, 3, figsize=(15, 6))
stats.probplot(semanal_lluvia['Promedio_Semanal'], dist="norm", plot=axs[0])
axs[0].set_title('Q-Q Precipitación')
# axs[0].set_ylim(0,1)

stats.probplot(semanal_rsolar['Promedio_Semanal'], dist="norm", plot=axs[1])
axs[1].set_title('Q-Q Radiación Solar')
# axs[1].set_ylim(0,1)

stats.probplot(semanal_tseca['Promedio_Semanal'], dist="norm", plot=axs[2])
axs[2].set_title('Q-Q Temperatura')
# axs[2].set_ylim(0,1)

plt.tight_layout()
plt.show()
```

#### Pruebas de Normalidad
la prueba de Shapiro-Wilk o la prueba de Kolmogorov-Smirnov pueden determinar si los datos se desvían significativamente de la normalidad.

```python
from scipy.stats import shapiro, kstest

# Shapiro-Wilk Test
print("Shapiro-Wilk Test:")
for var, data in zip(['Precipitación', 'Radiación Solar', 'Temperatura'], 
                     [semanal_lluvia, semanal_rsolar, semanal_tseca]):
    stat, p_value = shapiro(data['Promedio_Semanal'])
    print(f"{var}: Estadístico = {stat:.4f}, p-value = {p_value:.4f}")

# Kolmogorov-Smirnov Test
print("\nKolmogorov-Smirnov Test:")
for var, data in zip(['Precipitación', 'Radiación Solar', 'Temperatura'], 
                     [semanal_lluvia, semanal_rsolar, semanal_tseca]):
    stat, p_value = kstest(data['Promedio_Semanal'], 'norm', 
                           args=(data['Promedio_Semanal'].mean(), data['Promedio_Semanal'].std()))
    print(f"{var}: Estadístico = {stat:.4f}, p-value = {p_value:.4f}")
```

Los datos *'p-value' < 0.05* indican que los datos no siguen una distribución normal

## Análisis Bivariado: Correlación
El análisis bivariado se enfoca en estudiar la relación entre dos variables para entender cómo una variable puede afectar o estar relacionada con otra. La correlación es de los métodos mas utilizados y el que implementaremos para este proyecto.

```python
# Para hallar las medidas de correlación debemos ajustar todos los dataframe de una fecha que hayan datos de los 3
# hasta la fecha que termine el primero de los 3
f_inicio = '2014-12-29'
f_inicio = pd.to_datetime(f_inicio)
f_fin = '2020-01-31'
f_fin = pd.to_datetime(f_fin)

unificado_lluvia = semanal_lluvia[(semanal_lluvia['Fecha_Inicial'] >= f_inicio) & (semanal_lluvia['Fecha_Inicial'] <= f_fin)].copy()
print(unificado_lluvia)
unificado_rsolar = semanal_rsolar[(semanal_rsolar['Fecha_Inicial'] >= f_inicio) & (semanal_rsolar['Fecha_Inicial'] <= f_fin)].copy()
print(unificado_rsolar)
unificado_tseca = semanal_tseca[(semanal_tseca['Fecha_Inicial'] >= f_inicio) & (semanal_tseca['Fecha_Inicial'] <= f_fin)].copy()
print(unificado_tseca)
```

### Medidas de Correlación

```python
fig = plt.figure(figsize=(15, 6))

# Gráfico de dispersión entre lluvia y radiación solar
ax1 = fig.add_subplot(1, 3, 1)
ax1.scatter(unificado_lluvia['Promedio_Semanal'], unificado_rsolar['Promedio_Semanal'], color='blue', alpha=0.7)
ax1.set_title('Lluvia vs Radiación Solar')
ax1.set_xlabel('Lluvia (Promedio Semanal)')
ax1.set_ylabel('Radiación Solar (Promedio Semanal)')

# Gráfico de dispersión entre radiación solar y temperatura
ax2 = fig.add_subplot(1, 3, 2)
ax2.scatter(unificado_rsolar['Promedio_Semanal'], unificado_tseca['Promedio_Semanal'], color='orange', alpha=0.7)
ax2.set_title('Radiación Solar vs Temperatura')
ax2.set_xlabel('Radiación Solar (Promedio Semanal)')
ax2.set_ylabel('Temperatura (Promedio Semanal)')

# Gráfico de dispersión entre lluvia y temperatura
ax3 = fig.add_subplot(1, 3, 3)
ax3.scatter(unificado_lluvia['Promedio_Semanal'], unificado_tseca['Promedio_Semanal'], color='green', alpha=0.7)
ax3.set_title('Lluvia vs Temperatura')
ax3.set_xlabel('Lluvia (Promedio Semanal)')
ax3.set_ylabel('Temperatura (Promedio Semanal)')

plt.tight_layout()
plt.show()
```

```python
# Calculamos los coeficientes de correlación de Pearson
correlacion_pearson = unificado_lluvia['Promedio_Semanal'].corr(unificado_rsolar['Promedio_Semanal'])
print(f"Correlación Pearson entre lluvia y radiación solar: {correlacion_pearson:.2f}")

correlacion_pearson = unificado_rsolar['Promedio_Semanal'].corr(unificado_tseca['Promedio_Semanal'])
print(f"Correlación Pearson entre radiación solar y temperatura: {correlacion_pearson:.2f}")

# Calculamos los coeficientes de Spearman (opcional si no son relaciones lineales)
correlacion_spearman = unificado_lluvia['Promedio_Semanal'].corr(unificado_rsolar['Promedio_Semanal'], method='spearman')
print(f"Correlación Spearman entre lluvia y radiación solar: {correlacion_spearman:.2f}")

correlacion_spearman = unificado_rsolar['Promedio_Semanal'].corr(unificado_tseca['Promedio_Semanal'], method='spearman')
print(f"Correlación Spearman entre radiación solar y temperatura: {correlacion_spearman:.2f}")
```

La correlación entre los datos es menor a 0.5 lo que significa que si existe pero es una correlación baja.

## Análisis Multivariado: Análisis de Componentes Principales (PCA)
El análisis multivariado se utiliza para entender relaciones entre múltiples variables y reducir la complejidad de los datos. Uno de los métodos más importantes en el análisis multivariado es el Análisis de Componentes Principales (PCA)

- Componente Principal 1 (PC1): Captura la mayor parte de la varianza en los datos.
- Componente Principal 2 (PC2): Captura la siguiente mayor parte de la varianza, y es ortogonal a PC1.
- Componente N (PCN): Cada componente adicional captura una menor proporción de la varianza restante.

Proceso de PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

# Combinamos las columnas relevantes en un único DataFrame para el PCA
datos_pca = pd.DataFrame({
    'Lluvia': unificado_lluvia['Promedio_Semanal'],
    'Radiación Solar': unificado_rsolar['Promedio_Semanal'],
    'Temperatura': unificado_tseca['Promedio_Semanal']
})

# Estandarizamos los datos
scaler = StandardScaler()
datos_estandarizados = scaler.fit_transform(datos_pca)
# datos_estandarizados
```

```python
# Calculamos la matriz de covarianza
matriz_covarianza = pd.DataFrame(datos_estandarizados).cov()
print("Matriz de Covarianza:\n", matriz_covarianza)
```
