```python
import pandas as pd
import numpy as np
archivo_lluvia = 'datos_lluvia.csv'
datos_lluvia = pd.read_csv(archivo_lluvia)
df_lluvia = pd.DataFrame(datos_lluvia)
pd.set_option('display.max_rows', 6)
df_lluvia
```

Podemos darnos cuenta que contamos con datos desde el 2013-07-22 hasta el 2024-11-21

```python
archivo_rsolar = 'datos_rsolar.csv'
datos_rsolar = pd.read_csv(archivo_rsolar)
df_rsolar = pd.DataFrame(datos_rsolar)
df_rsolar
```

Hay datos de la radiación solar horaria desde 2014-12-08 hasta 2022-09-11

```python
archivo_tseca = 'datos_temp_seca.csv'
datos_tseca = pd.read_csv(archivo_tseca)
df_tseca = pd.DataFrame(datos_tseca)
df_tseca
```

Hay datos de la temperatura seca del aire cada 3 horas desde 1983-02-01 hasta 2020-01-31. Se tomó este dato ya que haciendo la respectiva investigación, ya que descubrimos que este es el dato generalmente usado para los pronosticos del clima en televisión y que muestran las aplicaciones

Todos los datos son tomados de la misma estación meteorológica, el aeropuerto internacional matecaña. Por lo tanto podemos ignorar muchos parámetros y enfocarnos en las variables como los campos ('VALOR') junto a sus respectivas fecha y hora para irlos clasificando

```python
campos = ['Fecha', 'Valor']
diaria_lluvia = df_lluvia[campos].copy()
diaria_lluvia['Fecha'] = pd.to_datetime(diaria_lluvia['Fecha'])
diaria_lluvia.rename(columns={'Valor': 'Precipitacion'}, inplace=True)
diaria_lluvia['Fecha'] = pd.to_datetime(diaria_lluvia['Fecha']) #Convierto la fecha en tipo datetime
```

```python
new_rsolar = df_rsolar[campos].copy()
new_rsolar['Fecha'] = pd.to_datetime(new_rsolar['Fecha'])
new_rsolar
```

```python
# Como los datos de radiación solar están cada hora están por horas debemos sacar el promedio de radiación solar por dia
new_rsolar = new_rsolar[new_rsolar['Valor'] >= 5].copy() #Solo me interesan las horas reales de sol durante el dia, de noche se toman como cero
new_rsolar.loc[:, 'Dia'] = new_rsolar['Fecha'].dt.date #Extraigo la fecha sin hora en una nueva columna

diaria_rsolar = new_rsolar.groupby('Dia').agg(
    Radiacion_Promedio=('Valor', 'mean')  # Calcula el promedio diario
).reset_index()

diaria_rsolar.rename(columns={'Dia': 'Fecha'}, inplace=True)
diaria_rsolar['Fecha'] = pd.to_datetime(diaria_rsolar['Fecha']) #Convierto la fecha en tipo datetime
```

```python
new_tseca = df_tseca[campos].copy()
new_tseca['Fecha'] = pd.to_datetime(new_tseca['Fecha'])
new_tseca
```

```python
new_tseca.loc[:, 'Dia'] = new_tseca['Fecha'].dt.date #Extraigo la fecha sin hora en una nueva columna
new_tseca

diaria_tseca = new_tseca.groupby('Dia').agg(
    Temperatura_Promedio=('Valor', 'mean')  # Calcula el promedio diario
).reset_index()

diaria_tseca.rename(columns={'Dia': 'Fecha'}, inplace=True)
diaria_tseca['Fecha'] = pd.to_datetime(diaria_tseca['Fecha']) #Convierto la fecha en tipo datetime
```

## Análisis de los gráficos
### Gráficos por año

```python
import matplotlib.pyplot as plt

# Gráfico de líneas
fecha_inicio = '2013-07-22'
fecha_inicio = pd.to_datetime(fecha_inicio) 

# Sólo gráficaremos después de fecha_inicio
lluvia_filtered = diaria_lluvia[diaria_lluvia['Fecha'] >= fecha_inicio].copy()
rsolar_filtered = diaria_rsolar[diaria_rsolar['Fecha'] >= fecha_inicio].copy()
tseca_filtered = diaria_tseca[diaria_tseca['Fecha'] >= fecha_inicio].copy()

plt.figure(figsize=(15, 8)) #Tamaño del gráfico

# Grafica cada variable
plt.plot(lluvia_filtered['Fecha'], lluvia_filtered['Precipitacion'], label='Precipitación (mm)', color='blue', alpha=0.7)
plt.plot(rsolar_filtered['Fecha'], rsolar_filtered['Radiacion_Promedio'], label='Radiación Solar (Wh/m2)', color='orange', alpha=0.7)
plt.plot(tseca_filtered['Fecha'], tseca_filtered['Temperatura_Promedio'], label='Temperatura Promedio (°C)', color='red', alpha=0.7)

# Etiquetas de los ejes y titulo
plt.title(f'Comparación de Precipitación, Radiación Solar y Temperatura ({fecha_inicio.year} en adelante)', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Valores', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.5)
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

El tamaño de las gráficas no es muy analizable, por lo que lo mejor sería escalarlas

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() # Normalización

# Normaliza cada dataset por separado
lluvia_filtered.loc[:,'Precipitacion_Normalizada'] = scaler.fit_transform(lluvia_filtered[['Precipitacion']])
rsolar_filtered.loc[:,'Radiacion_Normalizada'] = scaler.fit_transform(rsolar_filtered[['Radiacion_Promedio']])
tseca_filtered.loc[:,'Temperatura_Normalizada'] = scaler.fit_transform(tseca_filtered[['Temperatura_Promedio']])

plt.figure(figsize=(14, 6))

plt.plot(lluvia_filtered['Fecha'], lluvia_filtered['Precipitacion_Normalizada'], label='Precipitación (Normalizada)', color='blue', alpha=0.7)
plt.plot(rsolar_filtered['Fecha'], rsolar_filtered['Radiacion_Normalizada'], label='Radiación Solar (Normalizada)', color='orange', alpha=0.7)
plt.plot(tseca_filtered['Fecha'], tseca_filtered['Temperatura_Normalizada'], label='Temperatura (Normalizada)', color='green', alpha=0.7)

plt.title(f'Comparación Normalizada de Precipitación, Radiación Solar y Temperatura después ({fecha_inicio.date()} en adelante)', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Valor Normalizado', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

Para poder visualizar mejor los datos y tener una mejor consistencia, se opta por graficar de forma anual y que los datos se recopilen de forma semanal.

```python
#Para la lluvia
lluvia_filtered.loc[:, 'Semana'] = ((lluvia_filtered['Fecha'] - pd.Timestamp(fecha_inicio)).dt.days // 7) # Cada semana tendrá un número único
semanal_lluvia = lluvia_filtered.groupby('Semana')['Precipitacion_Normalizada'].mean().reset_index() # Agrupamos número de semana y promedio
semanal_lluvia['Fecha_Inicial'] = pd.to_datetime(fecha_inicio) + pd.to_timedelta(semanal_lluvia['Semana'] * 7, unit='d') # número de la semana
semanal_lluvia.rename(columns={'Precipitacion_Normalizada': 'Promedio_Semanal'}, inplace=True) #renombrando columnas
print(semanal_lluvia)

#Para la radiación solar
rsolar_filtered.loc[:, 'Semana'] = ((rsolar_filtered['Fecha'] - pd.Timestamp(fecha_inicio)).dt.days // 7)
semanal_rsolar = rsolar_filtered.groupby('Semana')['Radiacion_Normalizada'].mean().reset_index()
semanal_rsolar['Fecha_Inicial'] = pd.to_datetime(fecha_inicio) + pd.to_timedelta(semanal_rsolar['Semana'] * 7, unit='d')
semanal_rsolar.rename(columns={'Radiacion_Normalizada': 'Promedio_Semanal'}, inplace=True)

#Para la radiación solar
tseca_filtered.loc[:, 'Semana'] = ((tseca_filtered['Fecha'] - pd.Timestamp(fecha_inicio)).dt.days // 7)
semanal_tseca = tseca_filtered.groupby('Semana')['Temperatura_Normalizada'].mean().reset_index()
semanal_tseca['Fecha_Inicial'] = pd.to_datetime(fecha_inicio) + pd.to_timedelta(semanal_tseca['Semana'] * 7, unit='d')
semanal_tseca.rename(columns={'Temperatura_Normalizada': 'Promedio_Semanal'}, inplace=True)

# Gráfica
plt.figure(figsize=(14, 6))
plt.plot(semanal_lluvia['Fecha_Inicial'], semanal_lluvia['Promedio_Semanal'], label='Precipitación (Normalizada)', color='blue', alpha=0.7)
plt.plot(semanal_rsolar['Fecha_Inicial'], semanal_rsolar['Promedio_Semanal'], label='Radiación Solar (Normalizada)', color='orange', alpha=0.7)
plt.plot(semanal_tseca['Fecha_Inicial'], semanal_tseca['Promedio_Semanal'], label='Temperatura (Normalizada)', color='green', alpha=0.7)
plt.title(f'Comparación Normalizada de Precipitación, Radiación Solar y Temperatura semanal ({fecha_inicio.date()} en adelante)', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Valor Normalizado', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
